MATPLOTLIB

Несколько графиков в одной системе координат

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
x=np.linspace(0,10,100)
y1=[sqrt(i) for i in x]
y2=x
plt.title("График корня из x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid() # Отбражение сетки
plt.plot(x,y1,x,y2)
```

Несколько отдельных графиков

```python
x = np.linspace(-10, 10, 50)
y1 = x
y2 = [i**2 for i in x]
# Построение графиков
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)# (столбцы,строки, индекс) для отображения
plt.plot(x, y1)               # построение графика
plt.title("y1 = x") # заголовок
plt.ylabel("y1", fontsize=16) # ось ординат
plt.grid()                # включение отображение сетки
plt.subplot(1, 2, 2)
plt.title("y2 = x^2")
plt.plot(x, y2)               # построение графика
plt.xlabel("x", fontsize=16)  # ось абсцисс
plt.ylabel("y2", fontsize=16) # ось ординат
plt.grid() 
```

Столбчатая диаграмма

```python
x = [3, 1, 3, 12, 2, 4, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 
plt.bar(x, y)
plt.title("Столбчатая диаграмма") 
plt.legend(["blue"])#Легенда
plt.show()
```

Графики с использованией Step,stem,stackplot и scatter

```python
x=np.arange(0,7)
y=x
# Определение места шага
#pre - у слева от х
#mid - y  в середине интервала
#post - y справа от x
where_set = "mid"
plt.title("Step(конфетка)")
plt.step(x, y, linewidth=2,marker='*', where=where_set)
plt.grid()
```

```python
x = np.arange(0, 20)
y1 = np.array([i**2+2*i for i in x])
y2 = np.array([i**2+4*i for i in x])
y3 = np.array([2*i for i in x])
labels = ["y1", "y2", "y3"]
fig, ax = plt.subplots()
plt.title("Stackplot")
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc='upper left')
```

```python
x=np.arange(0,5,0.25)
y=np.array([(0.3)*i**2 for i in x])
plt.title("Stem")
plt.stem(x,y,linefmt="r--",markerfmt="^",bottom=1)
```

```python
x=np.arange(0,10.5,0.1)
y=np.cos(x)
plt.title("Я волна")
plt.scatter(x,y)
```
